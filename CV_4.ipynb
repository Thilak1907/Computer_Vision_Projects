```python
import cv2
import torch
from google.colab.patches import cv2_imshow  # Import cv2_imshow

# Load a pre-trained YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5m')

# Load the image
image_path = '/content/stock-photo.jpg'

image = cv2.imread(image_path)

original = image.copy() # Assign a copy of the original image to original

# Perform inference on the image
results = model(image)

# Render results (bounding boxes and labels)
image_with_boxes = results.render()[0]

# Ensure both images have the same data type
original = original.astype(image_with_boxes.dtype)

side_by_side = cv2.hconcat([original, image_with_boxes])

# Display the side-by-side images using cv2_imshow
cv2_imshow(side_by_side)
```
